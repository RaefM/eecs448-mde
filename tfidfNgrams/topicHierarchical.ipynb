```python
# STEP 1: Normalize state membership to be a proper probability distribution
from math import isclose

# Replace with source of states if changing from topic
topics_df = pd.read_csv('aita_topics_5.csv', sep='\t').drop(['Unnamed: 0', 'id'], axis=1)
topics_arr = topics_df.to_numpy()
n_topics = 5
topic_names = ['T0', 'T1', 'T2', 'T3', 'T4']
# end block

for row in topics_arr:
    num_zero = n_topics - np.count_nonzero(row)
    row_sum = np.sum(row)
    if not isclose(1.0, row_sum, rel_tol=1e-05): 
        row[row == 0] += (1 - row_sum) / num_zero
        
topics_df_normalized = pd.DataFrame(topics_arr, columns=topic_names)
topics_df_normalized
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

np.random.seed(448)
def train_test_split_data(df):
    X = df.drop(['is_asshole', 'id'], axis=1)
    y = df.is_asshole.values
    return train_test_split(X, y, test_size=0.2, random_state=448)
```

```python
%%time
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import balanced_accuracy_score, accuracy_score

def train_on_df_topics(d2v_topic_df):
    params = {'C': [0.01, 0.1, 1, 10, 100], 'l1_ratio': [0, 0.2, 0.4, 0.6, 0.8, 1]}
    d2v_topic_clfs = []
    X_train_all, X_test_all, y_train, y_test = train_test_split_data(d2v_topic_df)
    
    X_train = X_train_all.drop(topic_names, axis=1)
    X_train_weights = X_train_all[topic_names]
    X_test = X_test_all.drop(topic_names axis=1)
    X_test_weights = X_test_all[topic_names]
    
    # train a classifier for each topic, fitting fuzzily to each point based on the
    # probability that it came from the corresponding cluster
    for topic in topic_names:
        logistic_classifier = GridSearchCV(
            LogisticRegression(penalty = 'elasticnet', class_weight = 'balanced', solver = 'saga'), 
            params,
            n_jobs = 4, 
            cv = 5, 
            verbose = 3,
        )
        logistic_classifier.fit(X_train, y_train, **{'sample_weight': X_train_weights[topic]})
        d2v_topic_clfs.append(logistic_classifier)
        
    return d2v_topic_clfs, X_test, X_test_weights, y_test

d2v_topic_clfs, X_test, X_test_weights, y_test = train_on_df_topics(d2v_topic_df)
```

```python
def topic_ensemble_performance(clfs, X_test, X_test_weights, y_test):
    # for each test set
    probs_under_each_topic = np.zeros(len(X_test))
    for clf, topic in zip(clfs, topic_names):
        log_probs = clf.predict_log_proba(X_test)
        log_weights = np.log(X_test_weights[topic].to_numpy())
        # log(p(X=1|t0)p(t0)) = log(p(X=1|t0)) + log(p(t0))
        probs_under_each_topic += (np.exp((log_probs.T[1]).T + log_weights))
        
    print(probs_under_each_topic)
    y_pred = [1 if prob > 0.5 else 0 for prob in probs_under_each_topic]
    
    print("Had balanced accuracy: " + str(balanced_accuracy_score(y_test, y_pred)))
    

topic_ensemble_performance(d2v_topic_clfs, X_test, X_test_weights, y_test)
```
