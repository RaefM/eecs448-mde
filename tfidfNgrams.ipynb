```python
%load_ext autoreload
%config Completer.use_jedi = False
%autoreload 2
from preprocess import *
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
base_df = pd.read_csv('aita_preprocessed.csv', sep='\t')
base_df
```

```python
X = base_df.processed_body.values
```

```python
unigram_vectorizer = TfidfVectorizer(ngram_range=(1,1), max_df=0.75, min_df=2)
bigram_vectorizer = TfidfVectorizer(ngram_range=(1,2), max_df=0.75, min_df=2)
trigram_vectorizer = TfidfVectorizer(ngram_range=(1,3), max_df=0.75, min_df=2)
X_uni = data=unigram_vectorizer.fit_transform(X)
X_bi = data=bigram_vectorizer.fit_transform(X)
X_tri = data=trigram_vectorizer.fit_transform(X)
```

```python
uni_df = pd.DataFrame.sparse.from_spmatrix(X_uni)
bi_df = pd.DataFrame.sparse.from_spmatrix(X_bi)
tri_df = pd.DataFrame.sparse.from_spmatrix(X_tri)
```

```python
uni_df['id'] = base_df['id']
bi_df['id'] = base_df['id']
tri_df['id'] = base_df['id']
```

```python
uni_df.to_csv('aita_tfidf_unigram.csv', sep='\t', encoding='utf-8')
bi_df.to_csv('aita_tfidf_bigram.csv', sep='\t', encoding='utf-8')
tri_df.to_csv('aita_tfidf_trigram.csv', sep='\t', encoding='utf-8')
```
